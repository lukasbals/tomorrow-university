```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("data/Food_Production.csv", index_col=0) 
```

```python
display(df)
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='Land use per kilogram (m² per kilogram)', y=df.index, data=df)
plt.xlabel('Land use per kilogram (m² per kilogram)')
plt.ylabel('Food Product')
plt.title('Land Use per Kilogram of Food Product')
plt.show()
```

```python
fig, ax1 = plt.subplots(figsize=(12, 8))

color = 'tab:blue'
ax1.set_xlabel('Food Product')
ax1.set_ylabel('Freshwater withdrawals per kilogram (liters per kilogram)', color=color)
ax1.bar(df.index, df['Freshwater withdrawals per kilogram (liters per kilogram)'], color=color, alpha=0.6, label='Freshwater withdrawals per kilogram')
ax1.tick_params(axis='y', labelcolor=color)
ax1.set_xticklabels(df.index, rotation=90)

ax2 = ax1.twinx()  
color = 'tab:green'
ax2.set_ylabel('Land use per kilogram (m² per kilogram)', color=color)
ax2.plot(df.index, df['Land use per kilogram (m² per kilogram)'], color=color, marker='o', label='Land use per kilogram')
ax2.tick_params(axis='y', labelcolor=color)

fig.tight_layout()  
plt.title('Freshwater Withdrawals and Land Use per Kilogram of Food Product')
plt.show()
```
